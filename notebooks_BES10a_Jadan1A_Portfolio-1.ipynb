```python
#Matcha Stock and Strength Calculator
#This serves as calculator for stocks and used of matcha powder in a Matcha Café, with this it will be
#convenient and efficient to track the used and remaining stocks of the shop
```

```python
stock = float(input("Current matcha stock (grams): "))
drinks = float(input("Number of drinks to prepare: "))
strength = float(input("Matcha grams per drink: "))

```

```python
matcha_needed = drinks * strength
remaining_stock = stock - matcha_needed
```

```python
print("Matcha needed:", matcha_needed, "grams")
print("Remaining matcha stock:", remaining_stock, "grams")
```
